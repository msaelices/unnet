```python
from unnet.grad import Node
a = Node(2.0, name='a')
b = Node(3.0, name='b')
c = a * b
d = Node(1.5, name='d')
e = d + c + a
```

```python
e.backward()
```

```python
print(f'e.grad={e.grad}')
print(f'd.grad={d.grad}')
print(f'c.grad={c.grad}')
print(f'b.grad={b.grad}')
print(f'a.grad={a.grad}')
```

```python
from unnet.utils import draw
draw(e)
```
