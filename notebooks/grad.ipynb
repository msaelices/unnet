Let's create a bunch of nodes that store some values 

```python
from unnet.grad import Node

a = Node(2.0, name='a')
b = Node(3.0, name='b')
c = Node(1.5, name='c')
```

Now let's perform some mathematical operations to get a final result node

```python
d = a * b
e = d + c + a
print(e)
```

We can imagine that these kind of operations could apply to a neural network performing a forward propagation. Now let's calculate the gradients of all the nodes by executing the back propagation method

```python
e.backward()
```

We can print the gradients for all the nodes to see the calculated values

```python
print(f'e.grad={e.grad}')
print(f'd.grad={d.grad}')
print(f'c.grad={c.grad}')
print(f'b.grad={b.grad}')
print(f'a.grad={a.grad}')
```

Now, let's draw the resulting graph of operations, represented by a directed graph

```python
from unnet.utils import draw

draw(e)
```
